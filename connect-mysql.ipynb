```python
from sqlalchemy import create_engine

engine = create_engine('mysql+pymysql://testing1:testing1@localhost/testing')

with engine.connect() as conn:
    print("Kết nối thành công! ✅")
```

```python
excel_file = 'duong_dan_den_file_cua_ban.xlsx'
try:
    # 3. Đọc toàn bộ các tab trong file Excel
    # Cấu hình sheet_name=None sẽ trả về một dictionary dạng {tên_tab: DataFrame}
    all_sheets = pd.read_excel(excel_file, sheet_name=None)
    
    print("--- Bắt đầu quá trình load dữ liệu ---")
    
    # 4. Duyệt qua từng tab và ghi vào MySQL
    for sheet_name, df in all_sheets.items():
        table_name = sheet_name.strip()
        
        print(f"Đang ghi tab '{sheet_name}' vào bảng '{table_name}'...")
        
        df.to_sql(name=table_name, con=engine, if_exists='replace', index=False)
        
        print(f"-> Thành công! Đã load {len(df)} dòng vào bảng '{table_name}'.")
        
    print("\n--- Hoàn thành! Toàn bộ dữ liệu đã được nạp vào MySQL ---")
except Exception as e:
    print("Lỗi khi load file Excel hoặc ghi vào MySQL:", e)
    # 3. Đọc toàn bộ các tab trong file Excel
    # Cấu hình sheet_name=None sẽ trả về một dictionary dạng {tên_tab: DataFrame}
    all_sheets = pd.read_excel(excel_file, sheet_name=None)
    
    print("--- Bắt đầu quá trình load dữ liệu ---")
    
    # 4. Duyệt qua từng tab và ghi vào MySQL
    for sheet_name, df in all_sheets.items():
        # Xóa khoảng trắng thừa ở tên tab (nếu có) để tránh lỗi tên bảng trong MySQL
        table_name = sheet_name.strip()
        
        print(f"Đang ghi tab '{sheet_name}' vào bảng '{table_name}'...")
        
        # Ghi dữ liệu vào MySQL
        # if_exists='replace': Nếu bảng đã tồn tại sẽ xóa đi tạo lại. Bạn có thể đổi thành 'append' nếu muốn ghi chèn.
        # index=False: Không lưu cột chỉ mục (index) của pandas vào database.
        df.to_sql(name=table_name, con=engine, if_exists='replace', index=False)
        
        print(f"-> Thành công! Đã load {len(df)} dòng vào bảng '{table_name}'.")
        
    print("\n--- Hoàn thành! Toàn bộ dữ liệu đã được nạp vào MySQL ---")
```
