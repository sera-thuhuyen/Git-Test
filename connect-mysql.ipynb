```python
import pandas as pd
import random
import os
```

```python
from sqlalchemy import create_engine

engine = create_engine('mysql+pymysql://testing1:testing1@localhost/testing')

with engine.connect() as conn:
    print("Kết nối thành công! ✅")
```

```python
def load_excel_files_to_mysql(file_list, db_engine, mode='replace'):
    success_count = 0
    for file_path in file_list:
        if not os.path.exists(file_path):
            print(f"⚠️ Cảnh báo: Không tìm thấy file '{file_path}'. Bỏ qua.")
            continue
            
        file_name = os.path.basename(file_path)
        print(f"\n📂 Đang xử lý file: {file_name}")

        try:
            all_sheets = pd.read_excel(file_path, sheet_name=None)
            for sheet_name, df in all_sheets.items():
                table_name = sheet_name.strip()
                print(f"Đang ghi tab '{sheet_name}' vào bảng '{table_name}'...")
                df.to_sql(name=table_name, con=engine, if_exists='replace', index=False)
                print(f"-> Thành công! Đã load {len(df)} dòng vào bảng '{table_name}'.")
            
        except Exception as e:
            print(f"❌ Lỗi khi xử lý file {file_name}: {e}")
    print(f"\n🎉 HOÀN THÀNH! Đã xử lý xong {success_count}/{len(file_list)} file thành công.")
```

```python
excel_files_list = ['data\Fintech Projects Dataset.xlsx', 'data\World Flags Dataset Addition.xlsx']
load_excel_files_to_mysql(file_list=excel_files_list, db_engine=engine)
```
